```python
import numpy as np
import matplotlib.pyplot as plt
path='output/'
```

```python
hh = np.loadtxt(path + 'hh.dat')
time = np.loadtxt(path + 'time.dat')
theta = np.loadtxt(path + 'theta.dat')
uwind = np.loadtxt(path + 'uwind.dat')
vwind = np.loadtxt(path + 'vwind.dat')
Emis_IP = np.loadtxt(path + 'Emis_IP.dat')
Emis_MT = np.loadtxt(path + 'Emis_MT.dat')
conc = np.loadtxt(path + 'concentrations.dat')
```

```python
print(conc)
```

```python
time_index=[0,4,8,12,16,20]
Marker_plot=['o','v','^','s','x','*']
p1=plt.figure(dpi=300,figsize=(8,15))
ax1 = plt.axes([0.08,0.65,0.9,0.27]) # left botton width heigth
plt.setp(ax1,xlim=[0,16],ylim=[0,3000])
for i in range(6):
    plt.plot(uwind[time_index[i],:],hh,linewidth = 1,Marker=Marker_plot[i],Markersize=5)
plt.grid()
plt.legend(['0 h','4 h ','8 h ','12 h ','16 h ','20 h '])
ax1.set_xlabel('u (m/s)')
ax1.set_ylabel('h (m)')
ax1.set_title('Day 1')

ax2 = plt.axes([0.08,0.35,0.9,0.27])
plt.setp(ax2,xlim=[-4,8],ylim=[0,3000])
for i in range(6):
    plt.plot(vwind[time_index[i],:],hh,linewidth = 1,Marker=Marker_plot[i],Markersize=5)
plt.grid()
plt.legend(['0 h','4 h ','8 h ','12 h ','16 h ','20 h '])
ax2.set_xlabel('v (m/s)')
ax2.set_ylabel('h (m)')

ax3 = plt.axes([0.08,0.05,0.9,0.27])
plt.setp(ax3,xlim=[16,30],ylim=[0,3000])
for i in range(6):
    plt.plot(theta[time_index[i],:]-273.0,hh,linewidth = 1,Marker=Marker_plot[i],Markersize=5)
plt.grid()
plt.legend(['0 h','4 h ','8 h ','12 h ','16 h ','20 h '])
ax3.set_xlabel(r'$\theta (^oC)$')
ax3.set_ylabel('h (m)')
plt.savefig(path + 'Day1.png')
```
